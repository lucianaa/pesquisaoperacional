**Problema de Transporte**

Nesses tipos de problemas há uma decisão que envolve rotas e custas de forma a transportar diferentes itens entre cada fonte e cada destino.

Exemplo:

Existem três fontes de suprimento de um produto, indicadas por $F_1$, $F_2$ e $F_3$, com as seguintes capacidades: $F_1 = 10.000$ unidades, $F_2 = 15.000$ unidades, e $F_3 = 5.000$ unidades, que somam 30.000 unidades por mês. Essas fontes devem suprir as necessidades de 4 armazéns, indicados por $A_1$, $A_2$, $A_3$, e $A_4$, com as seguintes demandas: $A_1 = 8.000$ unidades, $A_2 = 4.000$ unidades, $A_3 = 7.000$ unidades, e $A_4 = 11.000$ unidades, totalizando 30.000 unidades também.

Associada a demanda e suprimento, temos os custos de transporte:

|      | $A_1$ | $A_2$| $A_3$ | $A_4$|
|--    | --    | --   | --    |  --  | 
|$F_1$ |13     | 8    |9      | 12   |
|$F_2$ | 12    | 9    | 10    | 14	 |
|$F_3$ |8      | 8    | 9     | 6    |


Para formular a situação como um problema de PL, chamemos de $x_{ij}$ o número de unidades do produto despachadas da fonte, com ${i=1,2,3}$ e para o destino com ${j=1,2,3,4}$. Então temos a nossa matriz de transporte:

|      | $A_1$ | $A_2$| $A_3$ | $A_4$|Suprimento|
|--    | --    | --   | --    |  --  | -- |
|$F_1$ |13 ($x_{11}$)|8 ($x_{12}$)|9 ($x_{13}$)|12($x_{14}$)   |10000|
|$F_2$ |12 ($x_{21}$)|9 ($x_{22}$)|10 ($x_{23}$)|14($x_{24}$)	 |15000|
|$F_3$ |8 ($x_{31}$) |8  ($x_{32}$)|9  ($x_{33}$)|6  ($x_{34}$)    |5000 |
|Demanda|8000  |4000  |7000   | 11000|30000|

Com 12 variáveis, agora é hora de modelar o problema. Como o objetivo é buscar o menor custo viável de transporte. O custo então será o produto da quantidade despachada por cada caminho pelo custo deste. Portanto, tempos

min $13 x_{11} + 8 x_{12} + 9 x_{13} + 12 x_{14} + 12x_{21} + 9 x_{22} + 10 x_{23} + 14 x_{24} + 8 x_{31} + 8 x_{32} + 9x_{33} + 6x_{34} $

Restrições relativas à produção das fontes:

$x_{11} + x_{12} + x_{13} + x_{14} \leq 10.000$

$x_{21} + x_{22} + x_{23} + x_{24} \leq 15.000$

$x_{31} + x_{32} + x_{33} + x_{34} \leq 5.000$

Restrições relativas às demandas:

$x_{11} + x_{21} + x_{31} = 8000$

$x_{12} + x_{22} + x_{32} = 4000$

$x_{13} + x_{23} + x_{33} = 7000$

$x_{14} + x_{24} + x_{34} = 11000$

A solução pode ser obtida por meio do método Simplex. A tabela completa é a seguinte:

|      | $A_1$ | $A_2$| $A_3$ | $A_4$|Suprimento|
|--    | --    | --   | --    |  --  | -- |
|$F_1$ |13 (0)|8 (0)|9 (4000)|12(6000)   |10000|
|$F_2$ |12 (8000)|9 (4000)|10 (3000)|14(0)	 |15000|
|$F_3$ |8 (0) |8  (0)|9  (0)|6  (5000)    |5000 |
|Demanda|8000  |4000  |7000   | 11000|30000|

O valor total do transporte é de 300.000 reais mensais.

Exercício: Considere o problema anterior, porém, altere a capacidade de suprimento das fontes para:  $F_1 = 12.000$ unidades, $F_2 = 15.000$ unidades, e $F_3 = 9.000$ unidades. Qual a solução desse novo problema?

**Problema da designação**

Consiste em 'escolher a melhor pessoa para uma determinada tarefa'. O objetivo do problema é determinar a designação de menor custo de trabalhadores a tarefas.

O elemento, na tabela abaixo, $x_{ij}$ (ou $c_{ii}$, encontrado na literatura), representa o custo de designar o trabalhador $i$ à tarefa /4j/4,

|      | $Tarefa_1$ | $Tarefa_2$| $Tarefa_3$ |
|--    | --    | --   | --    | 
|$Trabalhador_1$ |$x_{11}$    | $x_{12}$    |$x_{13}$       |
|$Trabalhador_2$ | $x_{21}$     | $x_{22}$     | $x_{23}$   |
|$Trabalhador_3$ |$x_{31}$       | $x_{32}$     | $x_{33}$  |

Como pode ser visto, o problema da designação é um caso especial do problema de transporte.

**Exemplo:**

Suponha que o Sr. Joe deseja contratar alguns serviços de jardinagem, de pintura de portas e de lavagem do carro da família. Solicitou três orçamentos de prestadores de serviços gerais e teve a seguinte proposta

|      | Jardinagem | Pinutra| Lavagem |
|--    | --    | --   | --    | 
|João |$15$    | $10$    |$9$   |
|Karen | $9$     | $15$     | $10$   |
|Terci |$10$       | $12$     | $8$  |

Qual a combinação que resulta em menor custo para o Sr. João?

Utilizaremos o **método húngaro** para resolver o problema, e ele consiste nas seguintes etapas:

1. Na matriz original, identifique o menor valor de cada linha e o subtraia de todas as entradas da respectiva linha.

|      | Jardinagem | Pinutra| Lavagem | Mínimo |
|--    | --    | --   | --    | --|
|João |$6$    | $1$    |$0$   | $9$ |
|Karen | $0$     | $6$     | $1$   | $9$ |
|Terci |$2$       | $4$     | $0$  | $8$ |

2. Na matriz resultante, identifique o menor valor de cada coluna e o subtraia de todas as entradas da respectiva coluna.

|      | Jardinagem | Pinutra| Lavagem | 
|--    | --    | --   | --    | 
|João |$6$    | $1$    |$0$   |
|Karen | $0$     | $6$     | $1$   | 
|Terci |$2$       | $4$     | $0$  | 
|Mínimo| 0| 1| 0 | 

A solução ótima pode ser identificada como a designação viável associada com os elementos zero da matriz.

|      | Jardinagem | Pinutra| Lavagem | 
|--    | --    | --   | --    | 
|João |$6$    | **$0$**    |**$0$**   |
|Karen | **$0$**     | $5$  | $1$  | 
|Terci |$2$       | $3$     | **$0$**  |
