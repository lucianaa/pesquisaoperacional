A programação linear é um modelo matemático popular para resolver problemas em que as variáveis possam ser expressas por meio de equações e/ou inequações lineares. São caracteríscas desse modelo (Moreira, 2010): <br/>
<ul>
    <li> um conjunto de variáveis que possam ser maximizado ou minimizado. Por exemplo, suponha a equação $4x + 3y$, representa função objetivo de um problema em que se deseja maximizar o lucro de fabricação de dois produtos $x$ e $y$.
    </li>
    <li> um conjunto de restrições, que se refere a alguns limites impostos a solução do problema.</li>
</ul>

Existem também algumas diretrizes para a formulação: os parâmetros e as variáveis de decisão. 
<br/> <br/>
Os parâmetros se referem aos valores pré-fixados, como por exemplo, os valores $4$ e $3$, na equação anterior, que indicam representam o lucro de dois produtos, e esse valores são fixos. 
<br/> <br/>
Já as variáveis de decisão podem assumir diversos valores que irá maximizar ou minimizar a função objetivo. No nosso exemplo, são as variáveis  $x$ e $y$, que podem representar a quantidade dos produtos $A$ e $B$.


**Exemplo 1:** Uma fábrica produz dois produtos, $A$ e $B$. Cada um deles deve ser processado por duas máquinas, $M_1$ e $M_2$. Devido à programação de outros produtos, que também utilizam essas máquinas, a máquina $M_1$ tem $24$ horas de tempo disponível para os produtos $A$ e $B$, enquanto a máquina $M_2$ tem $16$ horas de tempo disponível. Para produzir uma unidade do produto $A$, gastam-se 4 horas em cada máquina $M_1$ e $M_2$. Para produzir uma unidade do produto $B$, gastam-se $6$ horas na máquina $M_1$ e $2$ horas na máquina $M_2$. Cada unidade vendida do produto A gera um lucro de R\$ 80,0 e cada unidade vendida do produto B gera um lucro de R\$ 60,0. Existe uma previsão máxima de demanda para o produto B de 3 unidades, não havendo restrições quanto a demanda do produto A. Deseja-se saber quantas unidade de A e de B devem ser produzidas, de forma a maximizar o lucro e, ao mesmo tempo, obedecer a todas as restrições.


Função objetivo:  $80 x + 60y$ <br /> <br />
Restrições: <br /><br /> 
$4x+ 6y \leq 24$ <br />
$4x + 2y \leq 16$  <br />
$y \leq 3$  <br />
$x \geq >=0$ e $y\geq0$ 
<br /> <br />
**Formulação completa:** <br /> <br />
Maximizar $80 x + 60y$ <br />

Sujeito a <br />
$4x+ 6y \leq 24$ <br />
$4x + 2y \leq 16$  <br />
$y \leq 3$  <br />
$x \geq 0$ e $y\geq0$ 

```python
''' Método tentativa e erro
Considere  x=5 e y=2
'''
#Verificando as restrições
x=5
y=2
r1 = 4*x + 6*y
if r1<=24:
    print('Restrição atendida. Valor =', r1)    
else:
    print('Restrição violada. Valor = ', r1)

    
```

```python
''' Tentativa 2
Considere  x=2 e y=2
'''
#Verificando as restrições
x=2
y=2
r1 = 4*x + 6*y
if r1<=24:
    print('Restrição 1 atendida. Valor =', r1)    
else:
    print('Restrição 1 violada. Valor = ', r1)
    
r2=4*x + 2*y 
if r2<=16:
    print('Restrição 2 atendida. Valor =', r2)    
else:
    print('Restrição 2 violada. Valor = ', r2)

#y<=3
#x e y >=0
#Mas esse seria os valores que maximizam a função objetivo?
```

**Solução gráfica**

```python
'''Cada restrição será representada por uma reta para definição da região permissível de soluções'''
#na restrição 1, se y=0, então 4x=24 e x=6
#se x=0, então 0x + 6*y=24 e y=4

import matplotlib.pyplot as plt
import numpy as np
from matplotlib.patches import Polygon
from matplotlib.collections import PatchCollection

x1=[6, 0]
y1=[0, 4]
fig, ax=plt.subplots()

plt.plot(x1, y1, 'g-', linewidth=2, markersize=1)
plt.axis([0, 6, 0, 4])
```

```python
#na restrição 2, se y=0, então 4x=16 e x=4
#se x=0, então 2*y=16 e y=8

x2=[4, 0]
y2=[0, 8]

plt.plot(x1, y1, 'g-', x2,y2, 'r-',linewidth=2, markersize=1)
plt.axis([0, 6, 0, 8])
```

```python
#na restrição 3, se y<=3, 
x3=[0, 8]
y3=[3, 3]

plt.plot(x1, y1, 'g-', x2,y2, 'r-', x3,y3, 'b' ,linewidth=2, markersize=1)
plt.axis([0, 6, 0, 8])
plt.text(0 , 0, "P",horizontalalignment='center', fontsize=20)
plt.text(4 , 0, "Q",horizontalalignment='center', fontsize=20)
plt.text(0 , 3, "T",horizontalalignment='center', fontsize=20)
```

```python
#Como encontrar os demais vértices do poligono?
#RST
x3=[0, 8]
y3=[3, 3]

plt.plot(x1, y1, 'g-', x2,y2, 'r-', x3,y3, 'b' ,linewidth=2, markersize=1)
plt.axis([0, 6, 0, 8])
plt.text(0 , 0, "P",horizontalalignment='center', fontsize=20)
plt.text(4 , 0, "Q",horizontalalignment='center', fontsize=20)
plt.text(0 , 3, "T",horizontalalignment='center', fontsize=20)

#o ponto R é o encontro de
#4x+6y=24 
#4x+2y=12
#y=2 x=3

plt.text(3 , 2, "R",horizontalalignment='center', fontsize=20)

#o ponto S é o encontro de
#4x+6y=24 
#y=3
#y=3 x=3/2
plt.text(3/2 , 3, "S",horizontalalignment='center', fontsize=20)
```

```python
#Qual a solução ótima?
#Para o ponto P
x=0
y=0
f1 = 80*x+60*y
print('R:', f1)

#Para o ponto Q
x=4
y=0
f2 = 80*x+60*y
print('Q:', f2)

#Para o ponto R
x=3
y=2
f3 = 80*x+60*y
print('R:', f3)

x=3/2
y=3
f4 = 80*x+60*y
print('S:', f4)

x=0
y=3
f5 = 80*x+60*y
print('R:', f5)
```
